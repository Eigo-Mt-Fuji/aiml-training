## 1.2 単純パーセプトロン

### 1.2.1 単純パーセプトロンの概要

ニューラルネットの基本単位となる、複数の入力を受け取り1つの値（0 or 1）を出力するニューロンモデルを 単純パーセプトロン といいます。  
1.1.1の図のように、N入力1出力の単純パーセプトロンを数式に表すと以下のようになります。  

$$
\begin{eqnarray}
u = \sum_{i=1}^N w_i x_i + \theta\\
y = H(u)
\end{eqnarray}
$$

ここで、$H(u)$はuが0より大きいとき1、0以下のとき0となる関数で、ステップ関数といいます。  
$H$のように入力に対して処理を行なう関数を、活性化関数といいます。これについては1.3.5で詳しく扱います。  
上の式はそれぞれの入力$x_i$に重みと呼ばれる$w_i$をかけたものを合計し、バイアス$\theta$を足しています。  
実際のニューロンもこのモデルのように複数の入力を受けとり、それがしきい値に達すると発火とよばれる現象を起こすと考えられています。

### 1.2.2 単純パーセプトロンの実装

ここからは簡単な例として、2入力1出力の単純パーセプトロンについて考えます。 



この単純パーセプトロンを数式に表すと以下のようになります。
$$
\begin{eqnarray}
u = w_1 x_1 +  w_2 x_2 + \theta\\
y = H(u)
\end{eqnarray}
$$

また、ベクトルを用いて入力を${\bf x}=(x_{1}, x_{2})$, 重みを${\bf w}=(w_{1}, w_{2})$, 出力を$y$とすると次式のように簡潔にかけます。
$$
\begin{eqnarray}
        y = H({\bf w}^{\mathrm{T}}{\bf x} + \theta)
\end{eqnarray}
$$

### 1.2.3 論理関数の表現Ⅰ

1.2.1 で作成した単層パーセプトロンは$w$や$\theta$を変更することにより入力に対する反応が変化します。これを用いていくつかの関数を表現することができます。  
今回は2入力1出力の単純パーセプトロンを用いてAND関数を表現します。AND関数とは、2つの入力に対して以下のような出力をする関数です。  

|入力1|入力2||出力|
|:-----:|:------:||:----:|
|0|0||0|
|0|1||0|
|1|1||1|
|1|0||0|

#### 問題

下のコードを実行するとグラフが描画されます。このモデルではAND関数を表現できていません。  
グラフを見ながらwとbを変更して、AND関数を表現する単純パーセプトロンを作成して下さい。

```python
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

# H(u)を複数の入力に対応させる
def H(u):
    return np.where(u > 0, 1, 0)


x = np.array([[0, 0], [0, 1], [1, 1], [1, 0]]).T

#---------ここを変更-----------------
w = np.array([-2, 2])
b = 1.5
#--------------------------------------

u = np.dot(w, x) + b
y = H(u)

# 出力の確認用
print('入力1 入力2 出力\n'
      '  0         0       {0[0]}  \n'
      '  0         1       {0[1]}  \n'
      '  1         1       {0[2]}  \n'
      '  1         0       {0[3]}  \n'.format(y))
print('w_1 = {}, w_2 = {}, b = {}\n'.format(w[0], w[1], b))

# プロット
x_plt = np.arange(-1, 2, 0.1)
y_plt = -(w[0] / w[1]) * x_plt - b / w[1]
plt.plot(x_plt, y_plt)
for xt in x.T:
    if H(np.dot(w, xt) + b) > 0:
        color = 'r'
    else:
        color = 'b'
    plt.scatter(xt[0], xt[1], s=50, c=color)
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.xlabel('x_1')
plt.ylabel('x_2')
print('赤い点は1、青い点は0を示します。')
plt.show()
```

#### ヒント  
1.1.3で示したステップ関数$H$の中身$u$は  
$${\bf w}^{\mathrm{T}}{\bf x}+\theta=w_1 x_1+w_2 x_2+\theta$$
であり、 u=0とu=1の境界となるx,wは次の式を満たします。
$$w_1 x_1+w_2 x_2+\theta=0$$
つまり$x_1$-$x_2$グラフ上での境界を表す直線の式は
$$x_2 = - \frac{w_1}{w_2}x_1-\frac{\theta}{w_2}$$
です。  これを踏まえて$w$と$\theta$を変更するとうまく分類ができます。

#### 解答例

```python
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

# H(u)を複数の入力に対応させる
def H(u):
    return np.where(u > 0, 1, 0)


x = np.array([[0, 0], [0, 1], [1, 1], [1, 0]]).T

#---------ここを変更-----------------
w = np.array([1, 1])
b = -1.5
#--------------------------------------

u = np.dot(w, x) + b
y = H(u)

# 出力の確認用
print('入力1 入力2 出力\n'
      '  0         0       {0[0]}  \n'
      '  0         1       {0[1]}  \n'
      '  1         1       {0[2]}  \n'
      '  1         0       {0[3]}  \n'.format(y))
print('w_1 = {}, w_2 = {}, b = {}\n'.format(w[0], w[1], b))

# プロット
x_plt = np.arange(-1, 2, 0.1)
y_plt = -(w[0] / w[1]) * x_plt - b / w[1]
plt.plot(x_plt, y_plt)
for xt in x.T:
    if H(np.dot(w, xt) + b) > 0:
        color = 'r'
    else:
        color = 'b'
    plt.scatter(xt[0], xt[1], s=50, c=color)
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.xlabel('x_1')
plt.ylabel('x_2')
print('赤い点は1、青い点は0を示します。')
plt.show()
```

### 1.2.4 誤り訂正学習

前セッションの問題のように、$w_1$、$w_2$、$\theta$を手作業で変更し求めるモデルを得ることは大変です。  
そこで誤り訂正学習という方法を用いて$w_1$、$w_2$、$\theta$の更新を自動化します。  
先に簡単のため${\bf w}$, ${\bf x}$を以下のように置き直します。  
$$
{\bf w} = (w_1, w_2, \theta)\\
{\bf x} = (x_1, x_2, 1)\\
$$
さらに$u$を次のように定義し直します。
$$
u = {\bf w}^{\mathrm{T}}{\bf x}\\
$$

このとき
$$
u = w_1 x_1 + w_2 x_2 + \theta\\
$$
であるので、前セッションまでと本質的には変わらず$\theta$を省略して表記できることがわかります。<br>  
入力と出力には0or1しか現れないという前提のとき、重み${\bf w}$に適当な（間違った）初期値を与えた上で  
次に示す誤り訂正学習法（ヘッブ学習則）を繰り返し用いて少しずつ重み${\bf w}$を更新しモデルに最適化させることができます。
1. モデルに入力${\bf x_i} == (x_{i1}, x_{i2}, 1)$を与え、出力$y_i$を得る  
2. 正しい出力$t_i$とモデルの出力$y_i$の差から、以下の式で${\bf w}$を更新する  
$$
{\bf w} = {\bf w} + \eta(t_i - y_i){\bf x_i}
$$

ここで$\eta$は学習率と呼ばれるもので、通常正の値で定めます。詳しくは1.3.4で学びます。  
$t_i$と${\bf x}_i$はそれぞれ i 番目の正解データと入力データ、$y_i$は${\bf x}_i$を入力したときのニューロンの出力です。  
式から分かるように$w_1$は、$t_i$と$y_i$の値が異なり、かつ$x_{i1}$が1のときのみ更新されます。  
また、$y_i$の方が正解$t_i$より大きいとき$w$は負の方向に更新され、$y_i$の方が正解$t_i$より小さいとき$w$は正の方向に更新されます。

#### 問題

以下のプログラムの${\bf w}$を更新する部分を埋めて自動でAND関数を学習させて下さい。

```python
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

def H(u):
    return np.where(u > 0, 1, 0)


x = np.array([[0, 0, 1],  [0, 1, 1], [1, 1, 1], [1, 0, 1]]).T
t = np.array([0, 0, 1, 0])
w = np.array([-2, 2, 1.5])
eta = 0.1

# 同じデータで50回学習する
for i in range(50):
    for j, x_in in enumerate(x.T):
        u = np.dot(w, x_in)
        y = H(u)
        #---------------------------
        #        ここを書いて下さい
        #---------------------------


u = np.dot(w, x)
y = H(u)

print('入力1 入力2 出力\n'
      '  0         0       {0[0]}  \n'
      '  0         1       {0[1]}  \n'
      '  1         1       {0[2]}  \n'
      '  1         0       {0[3]}  \n'.format(y))
print('w = ({0[0]:.3f}, {0[1]:.3f}, {0[2]:.3f})'.format(w))
```

#### 解答例

```python
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

def H(u):
    return np.where(u > 0, 1, 0)


x = np.array([[0, 0, 1],  [0, 1, 1], [1, 1, 1], [1, 0, 1]]).T
t = np.array([0, 0, 1, 0])
w = np.array([-2, 2, 1.5])
eta = 0.1

# 同じデータで50回学習する
for i in range(50):
    for j, x_in in enumerate(x.T):
        u = np.dot(w, x_in)
        y = H(u)
        #---------------------------
        w = w + eta * (t[j] - y) * x_in
        #---------------------------


u = np.dot(w, x)
y = H(u)

print('入力1 入力2 出力\n'
      '  0         0       {0[0]}  \n'
      '  0         1       {0[1]}  \n'
      '  1         1       {0[2]}  \n'
      '  1         0       {0[3]}  \n'.format(y))
print('w = ({0[0]:.3f}, {0[1]:.3f}, {0[2]:.3f})'.format(w))
```

### 1.2.5 単純パーセプトロンの難点

XOR関数は以下のような入出力関係があります。

|入力1|入力2||出力|
|:-----:|:------:||:----:|
|0|0||0|
|0|1||1|
|1|1||0|
|1|0||1|

これは単純パーセプトロンでは表現できません。上の問題と同じように実装して確認してみます。  

```python
import numpy as np

x = np.array([[0, 0, 1], [0, 1, 1], [1, 1, 1], [1, 0, 1]]).T
t = np.array([0, 1, 0, 1])
w = np.array([-2, 2, -1.5])
eta = 0.1

# 同じデータで50回学習する
for i in range(50):
    for j, x_in in enumerate(x.T):
        u = np.dot(w, x_in)
        y = H(u)
        w = w + eta * (t[j] - y) * x_in


u = np.dot(w, x)
y = H(u)

print('入力1 入力2 出力\n'
      '  0         0       {0[0]}  \n'
      '  0         1       {0[1]}  \n'
      '  1         1       {0[2]}  \n'
      '  1         0       {0[3]}  \n'.format(y))
print('w = ({0[0]:.3f}, {0[1]:.3f}, {0[2]:.3f})'.format(w))
```

## 1.3 多層パーセプトロン（MLP）

### 1.3.1 多層パーセプトロン（MLP）の概要

これまでに作成したのは、入力がすぐに出力に伝搬されるようなモデルでした。  
多層パーセプトロンはこれをいくつか重ねたモデルで、下図のようなものです。



出展:http://opencv.jp/opencv-1.0.0/document/opencvref_ml_nn.html


例として、下図のような隠れ層が1つの多層パーセプトロンについて考えます。



上の図は、

$$
{\bf x} = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix},\  
{\bf W1} = \begin{pmatrix} w1_{11} & w1_{12} \\ w1_{21} & w1_{22} \end{pmatrix},\  
{\bf \theta1} = \begin{pmatrix} \theta1_1 \\ \theta1_2 \end{pmatrix},\  
{\bf u1} = \begin{pmatrix} u1_1 \\ u1_2 \end{pmatrix},\  
{\bf z1} = \begin{pmatrix} z1_1 \\ z1_2 \end{pmatrix},\  
{\bf w2} = \begin{pmatrix} w2_1 \\ w2_2 \end{pmatrix},\  
\theta2,\ 
u2,\ 
y\\
$$
を用いて以下のように表せます。
$$
{\bf u1} = {\bf W1}^{\mathrm{T}} {\bf x} + {\bf \theta1}\\
{\bf z1} = H({\bf u1})\\
u2 = {\bf w2}^{\mathrm{T}}{\bf z1} + \theta2\\
y = I(u2)
$$

```python
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

# H(u)を複数の入力に対応させる
def H(u):
    return np.where(u > 0, 1, 0)


x = np.array([[0, 0], [0, 1], [1, 1], [1, 0]]).T

w1 = np.array([1, 1])
b1 = -0.5
w2 = np.array([-1, -1])
b2 = 1.5
w3 = np.array([1,1])
b3 = -1.5

#---------------------------
u1 = np.dot(w1, x) + b1
u2 = np.dot(w2, x) + b2
#---------------------------
z1 = H(u1)
z2 = H(u2)
#---------------------------
u3 = np.dot(w3, np.array([z1,z2])) + b3
#---------------------------
y = H(u3)

# 出力の確認用
print('入力1 入力2 出力\n'
      '  0         0       {0[0]}  \n'
      '  0         1       {0[1]}  \n'
      '  1         1       {0[2]}  \n'
      '  1         0       {0[3]}  \n'.format(y))

# プロット
for (i, y_) in enumerate(y):
    if y_ > 0:
        color = 'r'
    else:
        color = 'b'
    plt.scatter(x[0,i], x[1,i], s=50, c=color)
plt.xlim([-0.1, 1.1])
plt.ylim([-0.1, 1.1])
plt.xlabel('x_1')
plt.ylabel('x_2')
print('赤い点は1、青い点は0を示します。')
plt.show()
```

### 1.3.3 勾配降下法

1.2の単純パーセプトロンの実装では、 以下のように重みを更新していました。
$$
{\bf w} \leftarrow {\bf w} + \eta(t_i - y_i){\bf x_i}
$$
しかし、実はこれでは入力が0or1以外の値をとる場合には対応できません（上式の場合、大きな$x$に対して$w$も大きくなってしまい誤差がより大きくなるから）。  
定量的に重みの更新量を評価するためには、ただし誤差関数を$E$とおいて次に示す勾配降下法という重みの更新方法を使います。
$$
{\bf w} \leftarrow {\bf w} - \eta \frac{dL}{d{\bf w}}
$$
隠れ層がどれだけ増えてもこの式で全ての重みを更新することができます。  
学習率$\eta$については1.3.2で、やや複雑ｂな微分ついては1.3.3で学びます。  
誤差関数には、二乗誤差などが使われます。  
モデルの学習では誤差関数$L$が小さくなるように重みを更新します。  

ここでは誤差関数が二次関数だったとして、誤差関数を最小化するよう勾配降下法で学習を行います。  
二次関数とその微分は下図のようなイメージです。



勾配降下法のイメージを次の図に示します。

### 1.3.4 学習率

学習率とは、勾配降下法での$\eta$のように、1ステップでどれだけ学習するかを調整するパラメーターです。  
学習率が大きいほど大幅に更新され、小さいほど更新量は小さくなります。  
学習率が大きすぎると、求めたい値に収束せずにむしろ発散してしまいます。小さい場合は移動回数が十分多いとき収束することが知られていますが、その分計算コストがかかります。 そのため、適切な学習率を決める必要があります。  
学習率は定数とは限らず、学習が進むにつれて変化させることもあります。 このことについてはChapter3で詳しく説明します。   

#### 問題

以下のプログラムのetaを変更して学習率の大きさが学習に与える影響を確認して下さい。

```python
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline

def quad(x):
    return 2 * x**2 + 5 * x + 1


def quad_deriv(x):
    return 4 * x + 5


x = 2
eta = 0.005
xs = np.array([x])
for i in range(50):
    x -= eta * quad_deriv(x)
    xs = np.append(xs, x)

x_plot = np.arange(-5.25, 2.85, 0.1)
y_quad = quad(x_plot)
ys = quad(xs)
plt.plot(x_plot, y_quad, label='quadratic curve')
plt.plot(xs, ys, label='trajectory')
plt.scatter(xs, ys)
plt.xlim([-5.25, 2.75])
plt.ylim([-3, 30])
plt.legend(loc='best')
plt.show()
```

#### 解答例

eta == 0.51


eta == 0.45


eta == 0.1  


etaを適切に設定しないとうまく学習が進まなくなってしまうので注意が必要です。  
etaの設定は基本的にはトライ＆エラーですが、ツールを使って適切な値を自動的に求めることもできます。

### 1.3.6 確率的勾配降下法

「同じ位置の勾配毎回は変わりません」を正しい表記にしましょう。 


ニューラルネットワークの重みの更新は誤差逆伝播により求まる誤差関数の勾配方向に更新されます。 このとき、一回の更新に用いられるデータの数によって呼ばれ方が異なります。  
名称は以下のとおりです。  

|名称|内容|
|:-:|:-|
|最急降下法|全データを用いて求めた勾配に基づいて更新|
|確率的勾配降下法|i番目のデータから求めた勾配で更新を$i=1, ..., N$に対して行う(Nはデータ数)|
|ミニバッチ学習|幾つかのデータに対して求めた勾配に基づいて更新を全データを使うまで行う|

一般的にどの方法でも、何度も学習させます。
この学習させる回数をエポック数といいます。

最急降下法は全データを用いるため、同じ位置の勾配は変わりません。そのため、一度局所解に収束してしまうと抜け出せなくなるという問題があります。  
これに対して、確率的勾配降下法では同じ位置でも勾配が変化するので、局所解に陥りにくくなります。  
ただし、1つのデータしか用いないため、それが外れ値であった場合は間違った方向に更新してしまいます。  
この問題を減らせるのがミニバッチ学習です。 ミニバッチ学習ではいくつかのデータを使うため、外れ値の影響が小さくなります。 このとき、1度に使うデータ数は自動では求まらないため決める必要があり、大きくしすぎると最急降下法に近くなり、小さくすると確率的勾配降下法に近くなります。
